## Is there a significant difference in median value of houses bounded by the Charles river or not? (T-test for independent samples)

A: MEDV Value CHAS is 1
B: MEDV Value CHAS is 0

NULL Hypothesis: Ho: A=B
Alternative Hypothesis: Ha: A!=B


```python
import scipy
```

```python
scipy.stats.ttest_ind(boston_df[boston_df['CHAS']==1]['MEDV'],
                      boston_df[boston_df['CHAS']==0]['MEDV'])
```

Since the p<0.05 so we reject the null hypothesis 
and it is proved that there is a difference between MEDV by the CHAS.

## 2. Is there a difference in Median values of houses (MEDV) for each proportion of owner occupied units built prior to 1940 (AGE)? (ANOVA)



1. NULL Hypothesis:
There is no difference between MEDV Value due to the Different Age group
2. Alternative Hypothesis:
There is difference between MEDV Value due to the Different Age group


```python
a=boston_df[boston_df['ageGroup']=="35 And Younger"]['MEDV']
b=boston_df[boston_df['ageGroup']=="Between 35 and 70"]['MEDV']
c=boston_df[boston_df['ageGroup']=="Greater Than 70"]['MEDV']

scipy.stats.f_oneway(a,b,c)
```

Since p-value<0.05 so we reject the null hypothesis it is proved that
"There is a difference between MEDV Value Due to the Different Age group"


## 3. Can we conclude that there is no relationship between Nitric oxide concentrations and proportion of non-retail business acres per town? (Pearson Correlation)

1. NULL Hypothesis: There is no relationship between NOX and INDUS
2. ALternative Hypothesis: There is a relationship Between NOX ansd INDUS

```python
scipy.stats.pearsonr(boston_df['NOX'],boston_df['INDUS'])
```

since p<0.05 so it is proved that there is coorelation of 0.76 between NOX and INDUS

## What is the impact of an additional weighted distance to the five Boston employment centres on the median value of owner occupied homes? (Regression analysis)

```python
import statsmodels.formula.api as smf
```

```python
formula="MEDV ~ DIS"
model=smf.ols(formula,data=boston_df).fit()
model.summary()
```
